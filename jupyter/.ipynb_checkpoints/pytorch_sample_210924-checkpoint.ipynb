```python
import torch.nn as nn
from torch.optim import SGD
import math
import numpy as np

import pandas as pd
```

```python
def mk_data_set(data,n_data,n_len):
    _x = []
    _t = []
    for i in range(n_data):
        x = data[i:i+n_len]
        t = data[n_len]
        _x.append(x), _t.append(t)
    return _x,_t

def gen_sin_wave(N_pi=50, N=10000):
    np.random.seed(2)
    _theta = np.linspace(0,2*np.pi*N_pi,N)
    _noise = np.random.randn(N)
    _t = np.sin(_theta)
    _x = _t + _noise*0.1
    return _x,_t


_x,_t = gen_sin_wave()
print(_x.shape, _t.shape)
df = pd.DataFrame()
df["x"] = _x
df["t"] = _t
df.iloc[:400,:].plot()
```
